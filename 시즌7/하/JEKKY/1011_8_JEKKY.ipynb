```python
depth = int(input())
for i in range(1, depth+1):
    for k in range((depth-i), 0, -1):
        print(" ",end="")
    for j in range(1, i*2):
        print("*", end="")
    print()
for i in range(depth-1, 0,-1):
    for k in range(0, (depth-i)):
        print(" ",end="")
    for j in range(i*2, 1, -1):
        print("*", end="")
    print()
```
